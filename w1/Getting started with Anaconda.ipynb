Anaconda is a distribution of the Python and R programming languages for scientific computing, that aims to simplify package management and deployment. The distribution includes data-science packages suitable for Windows, Linux, and macOS.

<center><h3>Anaconda installation in Windows</h3></center>

1. Visit https://www.anaconda.com/products/distribution.
2. Download the Installer.
3. Once the download is complete, run the installer.

4. The following window appears.<br>
<br>
5. Click on <b>Next</b> and then click on <b>I agree.</b><br>
6. In the following window choose the <b>Just Me</b> option.<br>
<br>
7. Click on <b>Next</b>.
8. Choose the directory where you want the distribution to install.
9. Click on <b>Next</b>.
10. At the Advanced Installation Options screen, I recommend that you do not check <b>Add Anaconda to my PATH environment variable</b>
11. Click on <b>Next</b> and run the installer.
12. Finish the installation.
13. Launch Anaconda Navigator.




<h3>Installing Dependencies</h3><br>
1. Open Anaconda Navigator.<br>
2. Launch CMD.exe prompt.<br>
3. To install Matplotlib, in the cmd window type <b>conda install matplotlib</b><br>
4. Do the same for Numpy and Pandas.
